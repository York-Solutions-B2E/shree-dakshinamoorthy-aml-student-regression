### Additional function to add each values as input to add new student record to the postgres table

```python

import psycopg2

def insert_student_record(row):

    print(row)

    connection = psycopg2.connect(
        host='localhost',
        database='postgres',
        user='postgres',
        password='password'
    )

    cursor = connection.cursor()

    try:
        
        sql = '''
            INSERT INTO student_grades(school, sex, age, address, famsize, pstatus, medu, fedu, mjob, fjob, reason, guardian,
            traveltime, studytime, failures, schoolsup, famsup, paid, activities, nursery, higher, internet, romantic, famrel,
            freetime, goout, dalc, walc, health, absences, g1, g2, g3)
            VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s,
            %s, %s, %s, %s, %s, %s, %s)
        '''
     
        cursor.execute(sql, row)

        connection.commit()
    except Exception as e:
        # Handle exceptions (e.g., print an error message)
        print(f"Error: {e}")
    finally:
    
        cursor.close()
        connection.close()

```
